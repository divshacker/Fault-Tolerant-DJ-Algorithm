```python
from qiskit import QuantumCircuit
from qiskit_ionq import GPIGate, GPI2Gate, MSGate
from math import pi
```

```python
native_3=QuantumCircuit(4,4)
# H gate 1

native_3.append(GPIGate(0), [1])



# H Gate 3


native_3.append(GPIGate(0), [3])

native_3.barrier()

# CX Gate 1 0

native_3.append(MSGate(0, 0,0.25), [1, 0])
native_3.append(GPI2Gate(0.5), [0])
native_3.append(GPI2Gate(0.5), [1])
native_3.append(GPI2Gate(0.25), [1])

# CX Gate 3 2

native_3.append(MSGate(0, 0,0.25), [3, 2])
native_3.append(GPI2Gate(0.5), [2])
native_3.append(GPI2Gate(0.5), [3])
native_3.append(GPI2Gate(0.25), [3])
native_3.barrier()

native_3.barrier()

# Z & S Gates
native_3.append(GPIGate(0),[0])
native_3.append(GPIGate(0.375),[0])
native_3.append(GPIGate(0),[1])
native_3.append(GPIGate(0.125),[1])
native_3.append(GPIGate(0),[2])
native_3.append(GPIGate(0.375),[2])
native_3.append(GPIGate(0),[3])
native_3.append(GPIGate(0.125),[3])
native_3.barrier()

# H gate 0
native_3.append(GPI2Gate(0.25), [0])
native_3.append(GPIGate(0), [0])
# H Gate 1
native_3.append(GPI2Gate(0.25), [1])
native_3.append(GPIGate(0), [1])
# H Gate 2
native_3.append(GPI2Gate(0.25), [2])
native_3.append(GPIGate(0), [2])
# H Gate 3
native_3.append(GPI2Gate(0.25), [3])
native_3.append(GPIGate(0), [3])
native_3.barrier()
native_3.draw('mpl')
```
