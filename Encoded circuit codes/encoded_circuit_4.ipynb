```python
from qiskit import QuantumCircuit
from qiskit_ionq import GPIGate, GPI2Gate, MSGate
from math import pi
```

```python
native_4=QuantumCircuit(4,4)
# H gate 1

native_4.append(GPIGate(0), [1])

# H Gate 3

native_4.append(GPIGate(0), [3])

native_4.barrier()

# CX Gate 1 0

native_4.append(MSGate(0, 0,0.25), [1, 0])
native_4.append(GPI2Gate(0.5), [0])
native_4.append(GPI2Gate(0.5), [1])
native_4.append(GPI2Gate(0.25), [1])

# CX Gate 3 2

native_4.append(MSGate(0, 0,0.25), [3, 2])
native_4.append(GPI2Gate(0.5), [2])
native_4.append(GPI2Gate(0.5), [3])
native_4.append(GPI2Gate(0.25), [3])
native_4.barrier()

# Y Gate 0


# Y Gate 2

native_4.barrier()

# Z & S Gates
native_4.append(GPIGate(0.5),[0])
native_4.append(GPIGate(0.25),[0])

native_4.append(GPIGate(0.5),[1])
native_4.append(GPIGate(0.25),[1])


native_4.barrier()

# H gate 0
native_4.append(GPI2Gate(0.25), [0])
native_4.append(GPIGate(0), [0])

# H Gate 1
native_4.append(GPI2Gate(0.25), [1])
native_4.append(GPIGate(0), [1])

# H Gate 2
native_4.append(GPI2Gate(0.25), [2])
native_4.append(GPIGate(0), [2])

# H Gate 3
native_4.append(GPI2Gate(0.25), [3])
native_4.append(GPIGate(0), [3])
native_4.barrier()


native_4.draw('mpl')
```
